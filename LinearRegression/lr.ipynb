```python
#Peform Simple Linear Regression,
#to do that we need to assign the feature variable which is TV, and the Prediction Variables which is sale

# Putting feature variable to X
X = advertising['TV']

# Putting response variable to y
y = advertising['Sales']

#now you need to split your data set to training and set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 , random_state=100)

#now we have X_train and X_test, y_train and y_test
```

```python
#if you print the shape of the array you will notice that it is only rows with no columns
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#however It is a general convention in scikit-learn that observations are rows, while features are columns.
#that is why we need to convert all our X values to columns

import numpy as np

X_train = X_train[:, np.newaxis]
X_test = X_test[:, np.newaxis]

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#now we are ready to run the Linear Regression Alg
# import LinearRegression from sklearn
from sklearn.linear_model import LinearRegression

# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LinearRegression()

# Fit the model using lr.fit()
lr.fit(X_train, y_train)
```

```python
# We know that the Liear Regressino Algorithme will find this formula
#y = c + m X
#we call m the coefficients and c the intercept
# to Print the intercept and coefficients
print(lr.intercept_)
print(lr.coef_)
```

```python
#now as we trained our model, lets use the test data to predict
y_pred = lr.predict(X_test)
```

```python
#now we have the predction in y_pred, lets compare it with the actual which is y_test
#we will plot both of them in graph
import matplotlib.pyplot as plt

#generate 60 index(which is the size of the test data)
c = [i for i in range(1,61,1)]

fig = plt.figure()
#plot the actual data, y_test, in blue
plt.plot(c,y_test, color="blue", linewidth=2.5, linestyle="-")
#print the prediction data, y_pred in red
plt.plot(c,y_pred, color="red",  linewidth=2.5, linestyle="-")

fig.suptitle('Actual and Predicted', fontsize=20)              
plt.xlabel('Index', fontsize=18)                               
plt.ylabel('Sales', fontsize=16)                       
```

```python
#now lets see the residual error, which is simply the difference between y_pred and y_test
#basically the error will be around the value 0,

c = [i for i in range(1,61,1)]
fig = plt.figure()

#plot the difference
plt.plot(c,y_test-y_pred, color="blue", linewidth=2.5, linestyle="-")

fig.suptitle('Error Terms', fontsize=20)             
plt.xlabel('Index', fontsize=18)                      
plt.ylabel('ytest-ypred', fontsize=16)               
```

lets now calculate the mean sqaured error, and r2.

mean squared error is the standard of calculating error in Linear regression



the output will be MSE = 7.9, this means that my model could not match 8% of the values

The R2 value is calculated by

1- calculate the mean of the predicted values (the mean of all Ys value)

2- calculate the distance between actual value and the mean

3- calculate the distance between the predicted value and the mean

4- then devide them



```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```
